# Subplotting

Sometimes we need more than one plot, we might want to compare 3 similar graphs or show an evolution.  
To do so we can create subplots in our `figure` object to create a **grid** of plots.

````python
wallet_2015 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234,74,530]}
wallet_2020 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234*4,74*7,530*1.3]}

fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 1 ÿµŸÅÿå 2 ÿπŸÖŸàÿØ

axs[0].bar(wallet_2015["currency"], wallet_2015["value_in_euro"])
axs[0].set_title("Wallet 2015")

axs[1].bar(wallet_2020["currency"], wallet_2020["value_in_euro"])
axs[1].set_title("Wallet 2020")

plt.tight_layout()
plt.show()
````

You can again use the `plt.subplots()`.

Remember the DRY principle? Don't Repeat Yourself: do not write 6 very similar lines of code. Be smarter!

<details>
  <summary>üí°Hint</summary>

  Inspect the `axes` object that `plt.subplots()` returns.
  It's a NumPy array of AxesSubplot, with a shape of 2 by 3. If you flatten it, you can easily loop over it to plot your charts.

  ```
  for i, ax in enumerate(axes.flatten()):
  ```

</details>

````python
wallets = pd.DataFrame({
    "bitcoin": [234, 234*4, 234*2, 234*3, 234*5, 234*6],
    "ethereum": [74, 74*7, 74*3, 74*2, 74*5, 74*4],
    "euro": [530, 530*1.3, 530*2, 530*1.5, 530*1.8, 530*2.2]
}, index=[2015, 2020, 2021, 2022, 2023, 2024])

fig, axes = plt.subplots(2, 3, figsize=(15, 8))

for i, ax in enumerate(axes.flatten()):
    ax.bar(wallets.columns, wallets.iloc[i])
    ax.set_title(f"Wallet {wallets.index[i]}")

plt.tight_layout()
plt.show()
````

````python
bitcoin = [3,3,4,5,5,6,11,12,13,14,16,14,18,15,19,18,19,15,10,8,9,9,9,12,13]
ethereum = [1,1,1,2,2,3,3,4,4,4,4,8,9,12,10,9,14,15,10,8,6,5,5,8,7]
euro = [7,7,7,7.6,7,7.8,7,7,7,7,7.6,7,7,7,7,6.8,6.9,7,7,7,7,7,7.2,7.2,7.3]

wallet_2015 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234,74,530]}
wallet_2020 = {"currency": ["bitcoin","ethereum","euro"], "value_in_euro": [234*4,74*7,530*1.3]}

fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Top-left: Bitcoin, Ethereum, Euro line plots
axes[0,0].plot(bitcoin, label="Bitcoin")
axes[0,0].plot(ethereum, label="Ethereum")
axes[0,0].plot(euro, label="Euro")
axes[0,0].set_title("Cryptocurrency Trends")
axes[0,0].legend()

# Top-right: Wallet 2015
axes[0,1].bar(wallet_2015["currency"], wallet_2015["value_in_euro"], color='skyblue')
axes[0,1].set_title("Wallet 2015")

# Bottom row: Wallet 2020 (spanning both columns)
ax_bottom = fig.add_subplot(2,1,2)  # spans the bottom row
ax_bottom.bar(wallet_2020["currency"], wallet_2020["value_in_euro"], color='lightgreen')
ax_bottom.set_title("Wallet 2020")

plt.tight_layout()
plt.show()
````
